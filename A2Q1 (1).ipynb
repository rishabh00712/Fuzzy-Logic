```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Defining the membership function for fuzzy set A (Low speed limit)
def mu_A(x):
    if x < 30:
        return 1
    elif 30 <= x < 50:
        return (50 - x) / (50 - 30)
    elif 50 <= x < 70:
        return (70 - x) / (70 - 50)
    else:
        return 0
```

```python
# Defining the membership function for fuzzy set B (High speed limit)
def mu_B(x):
    if x < 60:
        return 0
    elif 60 <= x < 80:
        return (x - 60) / (80 - 60)
    elif 80 <= x < 100:
        return (x - 80) / (100 - 80)
    else:
        return 1
```

```python
# Calculating membership values for A and B
x_values = np.arange(0, 110, 5)
mu_A_values = np.array([mu_A(x) for x in x_values])
mu_B_values = np.array([mu_B(x) for x in x_values])

# Performing the union operation
union_values = np.maximum(mu_A_values, mu_B_values)

# Performing the intersection operation
intersection_values = np.minimum(mu_A_values, mu_B_values)

# Performing the complement operation
complement_A_values = 1 - mu_A_values
complement_B_values = 1 - mu_B_values

# Performing the difference operation
difference_values = mu_A_values - mu_B_values
```

```python
# Plotting the results
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(x_values, mu_A_values, label='μ_A(x)')
plt.plot(x_values, mu_B_values, label='μ_B(x)')
plt.title('Membership Functions')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_values, union_values, label='Union (A v B)', color='g')
plt.title('Union of A and B')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x_values, intersection_values, label='Intersection (A ∧ B)', color='r')
plt.title('Intersection of A and B')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x_values, complement_A_values, label='Complement of A', linestyle='--')
plt.plot(x_values, complement_B_values, label='Complement of B', linestyle='--')
```

```python
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(x_values, complement_A_values, label='μ_A(x)')
plt.plot(x_values,complement_B_values, label='μ_B(x)')
plt.title('Complement')
plt.xlabel('Speed (x)')

plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_values, difference_values, label='Union (A v B)', color='g')
plt.title('difference of A and B')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
```
